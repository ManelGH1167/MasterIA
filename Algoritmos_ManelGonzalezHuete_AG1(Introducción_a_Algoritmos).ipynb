## Decorador python para medir el tiempo de ejecución

```python
#importar libreria
import time
import numpy as np
```

```python
# Decorador para medir el tiempo de ejecución
def medir_tiempo(func):
    def wrapper(*args, **kwargs):
        inicio = time.perf_counter()
        resultado = func(*args, **kwargs)
        fin = time.perf_counter()
        print(f"Tiempo de ejecución de '{func.__name__}': {fin - inicio:.6f} segundos")
        return resultado
    return wrapper
```

## Sucesión_de_Fibonacci

```python
#Sucesión_de_
#https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
#Calculo del termino n-simo de la suscesión de Fibonacci
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

print("Sucesión de Fibonacci para 5:", Fibonacci(5))
print("Sucesión de Fibonacci para 32:", Fibonacci(32))
print("Sucesión de Fibonacci para 35:", Fibonacci(35))
```

```python
# Si en lugar de la recursión, iteramos y guardamos los valores --> O(n)

#Calculo del termino n-simo de la suscesión de Fibonacci utilizando caché de todos los valores
def FibonacciIter(N:int):
    ListaFibo = [1,1]
    for i in range (2,N+1):
        ListaFibo.append(ListaFibo[i-1] + ListaFibo[i-2])

    return ListaFibo[i]

#Calculo del termino n-simo de la suscesión de Fibonacci optimizando caché, guardando sólo los 2 últimos valores
def FibonacciIter2(N:int):
    ListaFibo = [1,1]
    for i in range (2,N+1):
        temp = ListaFibo[1]
        ListaFibo[1] = ListaFibo[0] + ListaFibo[1]
        ListaFibo[0] = temp

    return ListaFibo[1]

# Cache completa
print("Calculo Fibonocci con caché complesta:")
print(FibonacciIter(5))
print(FibonacciIter(32))
print(FibonacciIter(35))

# Cache 2 últimos valores
print("\nCalculo Fibonocci con caché 2 últimos valores:")
print(FibonacciIter2(5))
print(FibonacciIter2(32))
print(FibonacciIter2(35))
```

## Devolución de cambio por técnica voraz

```python
@medir_tiempo
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION

cambio_monedas(143,[50,25,10,5,1])
```

```python
def es_prometedora(SOLUCION,etapa):
  # print("Solucion:", SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    # print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      # print("Comprobando diagonal de " + str(i) + " y " + str(j) + " con resultado:", end=(""))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) :
        # print("Diagonal")
        return False
      # print("Falso")

  return True
```

```python
def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  # Recorre las filas ...
  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      # print ("prometedora", etapa)
      if etapa == N-1:
        print(solucion)
        escribe(solucion)
        print()
      else:
        # print ("else-reinas")
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0
```

```python
#Sparse Matrix
# [(0,0)=0, (0,1)=5, (0.2)=4, (0,3)=5, (1,1)=0, (1,3)=2, (1,4)=3 .... ]

################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################
```

```python
#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )
```

### **Divide y Vencerás**

```python
def distanciaMin(aLista):
    """
    Busca la distancia mínima entre los números de una lista usando recursividad (divide y vencerás).

    Parámetros:
    - aLista: Lista de números.

    Retorna:
    - distancia mínima encontrada
    - indice1: índice para la posición 1 de la lista de números
    - indice2: índice para la posición 2 de la lista de números
    """
    def divide_venceras(inicio, final):
        # Condición de salida recursividad: Si la sublista tiene un solo elemento, no hay pares para comparar
        if final - inicio <= 1:
            return float('inf'), -1, -1

        # Buscamos la mitad de la lista para dividirla en 2
        mitad = (inicio + final) // 2
        # Calculamos cada mitad independientemente
        dist_izq, indice1_izq, indice2_izq = divide_venceras(inicio, mitad)
        dist_der, indice1_der, indice2_der = divide_venceras(mitad, final)

        # Obtenemos la distancia mínima (y sus índices) de las dos sublistas
        if dist_izq < dist_der :
            dist_min = dist_izq
            indice1_min, indice2_min = (indice1_izq, indice2_izq)
        else:
            dist_min = dist_der
            indice1_min, indice2_min = (indice1_der, indice2_der)

        # Creamos una lista con todos los valores para una comparación cruzada
        lista_completa = [(num, index) for index, num in enumerate(aLista[inicio:final])]
        lista_completa.sort()  # Ordenamos por valor para facilitar la búsqueda cruzada

        # Recorremos la lista (que al estar ordenada) podemos comparar pares cercanos
        # para deterctar si hay un mínimo diferente al encontrado.
        for i in range(len(lista_completa)-1):
            dist = abs(lista_completa[i + 1][0] - lista_completa[i][0])
            if dist < dist_min:
                dist_min = dist
                indice1_min, indice2_min = lista_completa[i + 1][1], lista_completa[i][1]

        return dist_min, indice1_min, indice2_min

    # No hay suficiente información
    if(len(aLista) < 2):
        return None

    return divide_venceras(0, len(aLista))
```

```python
# Ejemplo con divide y venceras
time1 = time.time()
resultado = distanciaMin(Lista1D)
# Calcula el tiempo transcurrido
timeTotal = time.time() - time1

print(f"La distancia mínima es de: {resultado[0]}, para los puntos {resultado[1]} y {resultado[2]}. Total tiempo: {timeTotal}")
print(f"La complejidad (big O), con divide y vencerás es de: O(n*log(n)) debido a la ordenación")
```

### Complejidad Algorítmica
**Tiempo:**

* Cada nivel de recursión realiza una fusión de listas que tiene costo O(n).

* Cada nivel de recursión realiza una ordenación de listas, las ordenaciones suelen tener un costo que tiene costo 𝑂(𝑛*log(𝑛)).

* Como hay 𝑂(log(𝑛)) recursiones debido a la división.

* Por lo tanto, la complejidad total es: 𝑂(𝑛*log(𝑛)^2)

**Espacio:**
El espacio requerido para manejar la lista y la pila de recursión es O(n)

### **Iterativa - Variación cruzada**

```python
posiciones = [0,0]
minimo = 999999999
time1 = time.time()
# Crear una lista de pares (valor, índice)
Lista1D_index = [(num, index) for index, num in enumerate(Lista1D)]

# Ordenar por valores. podríamos usar uno de los algoritmos desarrollados: mergeSort o quickSort, pero
# para la resolución del algoritmo no aporta
Lista1D_index.sort()

# Comparar elementos consecutivos en la lista ordenada
for i in range(1, len(Lista1D_index)):
    dist = abs(Lista1D_index[i - 1][0] - Lista1D_index[i][0])
    if dist < minimo:
        minimo = dist
        posiciones = [Lista1D_index[i - 1][1], Lista1D_index[i][1]]

time2 = time.time()
# Calcula el tiempo transcurrido
timeTotal = time2 - time1

print(f"La distancia mínima es de: {minimo}, entre los puntos {posiciones}. Total tiempo: {timeTotal}")
print(f"La complejidad (big O), por validación cruzada es de: O(n*log(n)) debido a la ordenación")
```

## **Lista de números de 2D/3D**
Mejora de los algoritmos para poder trabajar con lista de vectores de 2 dimensiones o 3 dimensiones

```python
import math

def distancia_euclidiana (punto1, punto2):
    """
    Calcula la distancia euclidiana entre dos puntos.

    Parámetros:
    - point1: Coordenadas del primer punto para un vector (x1, y1...).
    - point2: Coordenadas del segundo punto para un vector (x2, y2...).

    Retorna:
    - La distancia euclidiana entre dos puntos.
    """
    suma = 0
    for coord1, coord2 in zip(punto1, punto2):
        suma += (coord2 - coord1) ** 2

    return math.sqrt(suma)
```

### **Fuerza bruta**

Adaptación Fuerza bruta para utilizar puntos 1D, 2D o 3D

```python
def distanciaMin_fueza_bruta_(aLista):
    """
    Busca la distancia mínima entre los números de una lista mediante fuerza bruta.

    Parámetros:
    - aLista: Lista con los punto (independientemente de su dimensión)

    Retorna:
    - La distancia mínima entre dos puntos, dentro de una lista.
    """
    posiciones = [0,0]
    minimo = 999999999
    punto1D = False

    # Si la lista es de enteros (1D)
    if isinstance(aLista[0], int):
        punto1D = True

    # Recorremos los elementos de la lista
    for i in range(len(aLista)):
        # Dado el punto i como pivote, recorremos el resto de elementos de la lista
        for j in range(i+1, len(aLista)):
            # Si la lista es de enteros la distancia la calculamos por resta
            if(punto1D):
                distancia = abs(aLista[i]-aLista[j])
            # Si la lista es vectores calculamos la distancia euclidiana
            else:
                distancia = distancia_euclidiana(aLista[i],aLista[j])

            # Si las distancias son mínimas guardamos el valor y la posición
            if(distancia < minimo):
                minimo = distancia
                posiciones = [i,j]

    # Retornamos la distancia mínima y su posición
    return minimo, posiciones

Lista = Lista1D
time1 = time.time()
minimo, posiciones = distanciaMin_fueza_bruta_(Lista)
timeTotal = time.time() - time1 # Calcula el tiempo transcurrido
print(f"La distancia mínima es de: {minimo}, entre los puntos {posiciones}, con valores: {Lista[posiciones[0]]} y {Lista[posiciones[1]]}. Total tiempo: {timeTotal}")
print("---------------")
Lista = Lista2D
time1 = time.time()
minimo, posiciones = distanciaMin_fueza_bruta_(Lista)
timeTotal = time.time() - time1 # Calcula el tiempo transcurrido
print(f"La distancia mínima es de: {minimo}, entre los puntos {posiciones}, con valores: {Lista[posiciones[0]]} y {Lista[posiciones[1]]}. Total tiempo: {timeTotal}")
print("---------------")
Lista = Lista3D
time1 = time.time()
minimo, posiciones = distanciaMin_fueza_bruta_(Lista)
timeTotal = time.time() - time1 # Calcula el tiempo transcurrido
print(f"La distancia mínima es de: {minimo}, entre los puntos {posiciones}, con valores: {Lista[posiciones[0]]} y {Lista[posiciones[1]]}. Total tiempo: {timeTotal}")
print("---------------")
print(f"La complejidad (big O), por fuerza bruta es de: O(n^2)")
```

Adaptación Divide y Vencerás para utilizar puntos 1D, 2D o 3D

```python
from typing_extensions import AsyncGenerator
def distanciaMin_General(aLista):
    """
    Busca la distancia mínima entre los números de una lista usando recursividad (divide y vencerás).

    Parámetros:
    - aLista: Lista de números.

    Retorna:
    - distancia mínima encontrada
    - indice1: índice para la posición 1 de la lista de números
    - indice2: índice para la posición 2 de la lista de números
    """

    def divide_venceras_general(inicio, final, punto1D):

        def magnitud(vector):
            """
            Calcula el módulo (longitud) de un vector en 3D.
            """
            # En este caso los vectores son un tupla con la coodenada en la posición 0 y el índice en la posición 1
            return math.sqrt(sum(v**2 for v in vector[0]))

        # Condición de salida recursividad: Si la sublista tiene un solo elemento, no hay pares para comparar
        if final - inicio <= 1:
            return float('inf'), -1, -1

        # Buscamos la mitad de la lista para dividirla en 2
        mitad = (inicio + final) // 2
        # Calculamos cada mitad independientemente
        dist_izq, indice1_izq, indice2_izq = divide_venceras_general(inicio, mitad, punto1D)
        dist_der, indice1_der, indice2_der = divide_venceras_general(mitad, final, punto1D)

        # Obtenemos la distancia mínima (y sus índices) de las dos sublistas
        if dist_izq < dist_der :
            dist_min = dist_izq
            indice1_min, indice2_min = (indice1_izq, indice2_izq)
        else:
            dist_min = dist_der
            indice1_min, indice2_min = (indice1_der, indice2_der)

        # Creamos una lista con todos los valores para una comparación cruzada
        lista_completa = [(num, index) for index, num in enumerate(aLista[inicio:final])]
        # Ordenamos por valor para facilitar la búsqueda cruzada
        # En vectores mayores a 1D se debe utilizar la magnitud (módulo) para la ordenación
        if punto1D:
            lista_completa.sort()
        else:
            lista_completa = sorted(lista_completa, key=magnitud)

        # Recorremos la lista (que al estar ordenada) podemos comparar pares cercanos
        # para deterctar si hay un mínimo diferente al encontrado.
        for i in range(len(lista_completa)-1):
            # Si la lista es de enteros la distancia la calculamos por resta
            if punto1D:
                dist = abs(lista_completa[i + 1][0] - lista_completa[i][0])
            # Si la lista es vectores calculamos la distancia euclidiana
            else:
                dist = distancia_euclidiana(lista_completa[i + 1][0], lista_completa[i][0])
            if dist < dist_min:
                dist_min = dist
                indice1_min, indice2_min = lista_completa[i][1], lista_completa[i+1][1]

        return dist_min, indice1_min, indice2_min

    # No hay suficiente información
    if(len(aLista) < 2):
        return None

    # Si la lista es de enteros (1D)
    if isinstance(aLista[0], int):
        punto1D = True
    else:
        punto1D = False

    return divide_venceras_general(0, len(aLista), punto1D)
```
